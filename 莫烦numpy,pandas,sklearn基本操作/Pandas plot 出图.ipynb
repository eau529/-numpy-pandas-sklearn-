```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 创建一个Series 
# 这是一个线性的数据，我们随机生成1000个数据，
# Series 默认的 index 就是从0开始的整数，但是这里我显式赋值以便让大家看的更清楚

# 随机生成1000个数据
data = pd.Series(np.random.randn(1000),index=np.arange(1000))
 
# 为了方便观看效果, 我们累加这个数据
data.cumsum()

# pandas 数据可以直接观看其可视化形式
data.plot()

plt.show()

# 就这么简单，熟悉 matplotlib 的朋友知道如果需要plot一个数据，
# 我们可以使用 plt.plot(x=, y=)，把x,y的数据作为参数存进去，但是data本来就是一个数据，所以我们可以直接plot。 生成的结果就是下图：
```
